```python
import numpy as np
import pandas as pd
import seaborn as sns
sns.set_style('darkgrid')
sns.set_palette('PRGn_r')
sns.set_context('notebook')

df = pd.read_table('data/coffeechain.txt')
df.head(10)
```

```python
## Drop Records because it means to be meaningless in data modelling

df = df.drop(['Records'], axis=1)
df.head(10)
```

```python
## Separating data into numeric and categorical parts for cleaning and
## analysis

df_numeric = df.select_dtypes(include=np.number)
```

```python
## Here we see how data distributed in categorical data

df_category = df.drop(df_numeric.columns, axis=1)
df_category.head(10)
```

```python
## From table above, we can see that date has the same information for the
## first 10 rows, let's see how it distributes

sns.scatterplot(x='Date', y='Profit', data=df)
plt.xticks(rotation=90)
```

```python
## It seems like the data only collected monthly at 12th or 13th of the 
## last 3 months

dic = {'2012-10-01 00:00:00.000': 'OCT 12',
       '2012-11-01 00:00:00.000': 'NOV 12',
       '2012-12-01 00:00:00.000': 'DEC 12',
       '2013-10-01 00:00:00.000': 'OCT 13',
       '2013-11-01 00:00:00.000': 'NOV 13',
       '2013-12-01 00:00:00.000': 'DEC 13'}
df['Date'] = df['Date'].map(dic)
df_category['Date'] = df_category['Date'].map(dic)
```

```python
## Let's visualize the relationship of profit from other categorical variables

for column in df_category.columns:
    print(df_category[column].value_counts())
    sns.catplot(x=column, y='Profit', data=df, kind='bar')
    if (df_category[column].nunique() > 5):
        plt.xticks(rotation=90)
```

```python
## Let's create dummy variables for categorical data

df_processed = pd.DataFrame()
for cat in df_category.columns:
    temp_cat = pd.get_dummies(df[cat], prefix=cat, drop_first=True)
    df_processed = pd.concat([df_processed, temp_cat], axis=1)
```

```python
## Let's scale the numerical variable

from sklearn.preprocessing import StandardScaler
for num in df_numeric.columns:
    scaler = StandardScaler()
    scaler.fit(df[[num]])
    temp = scaler.transform(df[[num]])
    df_processed[num+'_scaled'] = temp
```

```python
df_processed.columns
```

```python
## Here we separate the predicted variables from other variables

X = df_processed.drop(['Cogs_scaled', 'Margin_scaled', 'Sales_scaled', 'Difference_scaled','Profit_scaled'], axis=1)
y_cogs = df_processed['Cogs_scaled']
y_profit = df_processed['Profit_scaled']
y_margin = df_processed['Margin_scaled']
y_sales = df_processed['Sales_scaled']
```

```python
## Profit
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y_profit, test_size=0.2, random_state=101)
```

```python
## Baseline Model
```

```python
from sklearn.metrics import mean_squared_error

y_null = np.zeros_like(y_test, dtype=float)
y_null.fill(y_train.mean())
baseline = np.sqrt(mean_squared_error(y_test, y_null))
print("baseline RMSE: ", baseline)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import KFold, cross_val_score


lm = LinearRegression()
cv = KFold(n_splits=5, random_state=1, shuffle=True)
mse_scores = cross_val_score(lm, X_train, y_train, cv=cv, scoring='neg_mean_squared_error')
mse_mean = (-mse_scores).mean()
print(mse_mean)
print('RMSE cross validation: ', np.sqrt(mse_mean))
```

```python
from sklearn.metrics import mean_absolute_error

lm.fit(X_train, y_train)
prediction = lm.predict(X_test)
print('MAE: {}'.format(mean_absolute_error(y_test, prediction)))
print('MSE: {}'.format(mean_squared_error(y_test, prediction)))
print('RMSE: {}'.format(np.sqrt(mean_squared_error(y_test, prediction))))
```

```python
## Regularization
from sklearn.linear_model import Lasso

lr1 = Lasso()
cv = KFold(n_splits=5, random_state=1, shuffle=True)
mse_scores = cross_val_score(lr1, X_train, y_train, cv=cv, scoring='neg_mean_squared_error')
mse_mean = (-mse_scores).mean()
print(mse_mean)
print('RMSE cross validation: ', np.sqrt(mse_mean))
```

```python
lr1.fit(X_train, y_train)
prediction = lr1.predict(X_test)
print('MAE: {}'.format(mean_absolute_error(y_test, prediction)))
print('MSE: {}'.format(mean_squared_error(y_test, prediction)))
print('RMSE: {}'.format(np.sqrt(mean_squared_error(y_test, prediction))))
```

```python
from sklearn.linear_model import Ridge

lr2 = Ridge()
cv = KFold(n_splits=5, random_state=1, shuffle=True)
mse_scores = cross_val_score(lr2, X_train, y_train, cv=cv, scoring='neg_mean_squared_error')
mse_mean = (-mse_scores).mean()
print(mse_mean)
print('RMSE cross validation: ', np.sqrt(mse_mean))
```

```python
lr2.fit(X_train, y_train)
prediction = lr2.predict(X_test)
print('MAE: {}'.format(mean_absolute_error(y_test, prediction)))
print('MSE: {}'.format(mean_squared_error(y_test, prediction)))
print('RMSE: {}'.format(np.sqrt(mean_squared_error(y_test, prediction))))
```

```python
n_estimators = [int(x) for x in np.linspace(start = 10, stop = 100, num = 10)]
max_features = ['auto', 'sqrt']
max_depth = [int(x) for x in np.linspace(2, 10, num = 10)]
max_depth.append(None)
min_samples_split = [2, 5, 10]
min_samples_leaf = [1, 2, 4]
bootstrap = [True, False]
grid_search = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf,
               'bootstrap': bootstrap}
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import RandomizedSearchCV

rf = RandomForestRegressor()
rf_random = RandomizedSearchCV(estimator = rf, param_distributions = grid_search, n_iter = 10, cv = 5, verbose = 2, random_state = 42, n_jobs = -1)
```

```python
rf_random.fit(X_train, y_train)
```

```python
rf_random.best_params_
```

```python
rf_best = RandomForestRegressor(**rf_random.best_params_)
```

```python
cv = KFold(n_splits=5, random_state=1, shuffle=True)
mse_scores = cross_val_score(rf_best, X_train, y_train, cv=cv, scoring='neg_mean_squared_error')
mse_mean = (-mse_scores).mean()
print(mse_mean)
print('RMSE cross validation: ', np.sqrt(mse_mean))
```

```python
rf_best.fit(X_train, y_train)
prediction = rf_best.predict(X_test)
print('MAE: {}'.format(mean_absolute_error(y_test, prediction)))
print('MSE: {}'.format(mean_squared_error(y_test, prediction)))
print('RMSE: {}'.format(np.sqrt(mean_squared_error(y_test, prediction))))
```
